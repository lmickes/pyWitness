# Tutorial
## Gaussian Signal Detection Theory Model

```python
from pyWitness.ToyModel import (
    GaussianSDTModel,
    plot_gaussians_with_criterion,
    plot_roc,
    plot_dprime_and_pauc,
)
import matplotlib.pyplot as plt
```

### Build the model

```python
model = GaussianSDTModel(mu_noise=0.0, sigma_noise=1.0,
                         mu_signal=2, sigma_signal=1.0)
criterion = 0.0
```

### Plot distributions and ROC curve

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
plot_gaussians_with_criterion(model, criterion, ax=axes[0])
plot_roc(model, criterion, ax=axes[1])
plt.tight_layout()
```

### Plot d' and pAUC

```python
plot_dprime_and_pauc(model)
```
